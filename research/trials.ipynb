```python
from langchain_community.document_loaders import PyMuPDFLoader

# Point to the new, clean book
file_path = "/Users/sadiya/Medical-Chatbot-Gen-AI/data/book.pdf"

loader = PyMuPDFLoader(file_path)
extracted_data = loader.load()

print(f"Total pages: {len(extracted_data)}")
print(f"Page 10 Snippet: {extracted_data[9].page_content[:500]}")
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Create the splitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)

# Break the 1,317 pages into smaller pieces
text_chunks = text_splitter.split_documents(extracted_data)

print(f"Total number of chunks: {len(text_chunks)}")
print(f"Sample chunk: {text_chunks[100].page_content}")
```

```python
from langchain.schema import Document

def get_minimal_docs(full_docs):
    minimal_docs = []
    for doc in full_docs:
        # Get the source path, or 'Unknown' if it's missing
        src = doc.metadata.get("source", "Unknown")
        
        # Create a brand new, clean Document object
        clean_doc = Document(
            page_content=doc.page_content,
            metadata={"source": src}
        )
        minimal_docs.append(clean_doc)
    return minimal_docs

# 1. Clean the metadata first
cleaned_data = get_minimal_docs(extracted_data)

# 2. Now split the CLEANED data into chunks
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
text_chunks = text_splitter.split_documents(cleaned_data)

print(f"Cleaned and split into {len(text_chunks)} chunks.")
print(f"Metadata check: {text_chunks[0].metadata}")
```
